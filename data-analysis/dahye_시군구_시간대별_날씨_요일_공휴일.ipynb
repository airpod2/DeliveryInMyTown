## 1.3 날짜별 요일 정보 데이터

```python
monday = pd.read_csv('/content/drive/MyDrive/freq_with_holiday.csv', parse_dates=['날짜'])
monday
```

```python
monday.drop(['Unnamed: 0', 'holiday','배달건수','읍면동','day_name','시간대','광역시도','시군구'], axis=1,inplace=True)
monday
```

```python
day_sample = monday.groupby(['날짜','요일']).count().reset_index()
day_sample
```

요일정보는 빠진 날짜를 복원하지 않았던 기존의 배달데이터를 기준으로 만들었기 때문에 배달데이터가 존재하지 않는 날은 요일 정보도 없다. 아래의 8개의 날짜를 추가로 넣어주었다.
* 빠진 날짜 찾는 법 : day_sample과 df를 mapping한 뒤 날짜가 nan인 날짜에 unique()취하면 찾을 수 있다.

```python
missing = [('2020-05-18', '화'),
('2021-03-27', '토'),
('2021-03-28', '일'),
('2021-03-29', '월'),
('2021-03-30', '화'),
('2021-03-31', '수'),
('2021-07-20', '화'),
('2021-07-21', '수')]

missing_day = pd.DataFrame(missing, columns = ['날짜',	'요일'])
missing_day['날짜'] = pd.to_datetime(missing_day['날짜'], format='%Y-%m-%d')
missing_day
```

```python
day_sample.info()
```

```python
missing_day.info()
```

```python
full_day_sample = pd.concat([day_sample,missing_day],ignore_index=True)
full_day_sample = full_day_sample.set_index('날짜')
full_day_sample
```

### 날짜 요일 mapping

```python
test = df.copy()
test 
```

```python
test['요일'] = test['날짜'].map(full_day_sample['요일'])
test
```

```python
test[test['요일'].isna()]
```
