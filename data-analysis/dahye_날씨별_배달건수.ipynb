```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/시군구_시간대별_날씨_요일.csv', parse_dates=['날짜'])
df
```

```python
df.drop(['holiday_name', '요일'],axis=1, inplace=True)
df
```

## 1.1 area2

```python
normal = df[(df['강수형태'] =='없음') & (df['기온']>=5) & (df['기온']<=25)]
normal
```

```python
normal_df = normal.groupby(['광역시도','시군구']).mean().reset_index()
normal_df
```

```python
rain =  df[(df['강수형태'] =='비') & (df['강수']>=5)]
rain
```

```python
rain_df = rain.groupby(['광역시도','시군구']).mean().reset_index()
rain_df
```

```python
snow = df[(df['강수형태'] =='눈') & (df['강수']>=2)]
snow
```

```python
snow_df =snow.groupby(['광역시도','시군구']).mean().reset_index()
snow_df
```

```python
hot = df[(df['강수형태'] =='없음') & (df['기온']>=30)]
hot
```

```python
hot_df =hot.groupby(['광역시도','시군구']).mean().reset_index()
hot_df
```

```python
cold = df[(df['강수형태'] =='없음') & (df['기온']<= 0)]
cold
```

```python
cold_df = cold.groupby(['광역시도','시군구']).mean().reset_index()
cold_df
```

```python
cold_df.describe()
```

```python
df_all = pd.merge(left = normal_df, right=rain_df, on=['광역시도','시군구'], how='left',suffixes=('_normal', '_rain'))[['광역시도','시군구','배달건수_normal','배달건수_rain']]
df_all
```

```python
df_all = pd.merge(left = df_all, right=snow_df, on=['광역시도','시군구'], how='left',suffixes=('_df_all', '_snow'))[['광역시도','시군구','배달건수_normal','배달건수_rain','배달건수']]
df_all
```

```python
df_all.rename(columns={'배달건수':'배달건수_snow'},inplace=True)
df_all
```

```python
df_all = pd.merge(left = df_all, right=hot_df, on=['광역시도','시군구'], how='left',suffixes=('_df_all', '_hot'))[['광역시도','시군구','배달건수_normal','배달건수_rain','배달건수_snow','배달건수']]
df_all
```

```python
df_all.rename(columns={'배달건수':'배달건수_hot'},inplace=True)
df_all
```

```python
df_all = pd.merge(left = df_all, right=cold_df, on=['광역시도','시군구'], how='left')[['광역시도','시군구','배달건수_normal','배달건수_rain','배달건수_snow','배달건수_hot','배달건수']]
df_all
```

```python
df_all.rename(columns={'배달건수':'배달건수_cold'},inplace=True)
df_all
```

```python
df_all = df_all.round(1)
df_all
```

```python
df_all.columns = ['광역시도',	'시군구',	'normal',	'rain',	'snow'	,'hot',	'cold']
df_all
```

```python
df_all.drop(df_all[df_all['광역시도']=='세종특별자치시'].index, inplace=True)
df_all
```

```python
df_all.to_csv('/content/drive/MyDrive/시군구_날씨별_평균배달_최종.csv',index=False, encoding="utf-8-sig")
```

## 1.2 area1

```python
normal = df[(df['강수형태'] =='없음') & (df['기온']>=5) & (df['기온']<=25)]
normal_df = normal.groupby(['광역시도']).mean().reset_index()
normal_df
```

```python
rain = df[(df['강수형태'] =='비') & (df['강수']>=5)]
rain_df = rain.groupby(['광역시도']).mean().reset_index()
snow = df[(df['강수형태'] =='눈') & (df['강수']>=2)]
snow_df =snow.groupby(['광역시도']).mean().reset_index()
hot = df[(df['강수형태'] =='없음') & (df['기온']>=30)]
hot_df =hot.groupby(['광역시도']).mean().reset_index()
cold = df[(df['강수형태'] =='없음') & (df['기온']<= 0)]
cold_df = cold.groupby(['광역시도']).mean().reset_index()
```

```python
df_all = pd.merge(left = normal_df, right=rain_df, on=['광역시도'], how='left',suffixes=('_normal', '_rain'))[['광역시도','배달건수_normal','배달건수_rain']]

df_all = pd.merge(left = df_all, right=snow_df, on=['광역시도'], how='left',suffixes=('_df_all', '_snow'))[['광역시도','배달건수_normal','배달건수_rain','배달건수']]
df_all.rename(columns={'배달건수':'배달건수_snow'},inplace=True)

df_all = pd.merge(left = df_all, right=hot_df, on=['광역시도'], how='left',suffixes=('_df_all', '_hot'))[['광역시도','배달건수_normal','배달건수_rain','배달건수_snow','배달건수']]
df_all.rename(columns={'배달건수':'배달건수_hot'},inplace=True)

df_all = pd.merge(left = df_all, right=cold_df, on=['광역시도'], how='left')[['광역시도','배달건수_normal','배달건수_rain','배달건수_snow','배달건수_hot','배달건수']]
df_all.rename(columns={'배달건수':'배달건수_cold'},inplace=True)

df_all = df_all.round(1)
df_all.columns = ['광역시도', 'normal', 'rain', 'snow' ,'hot', 'cold']

df_all
```

```python
df_all.to_csv('/content/drive/MyDrive/시군구_날씨별_평균배달_최종_area1.csv',index=False, encoding="utf-8-sig")
```
